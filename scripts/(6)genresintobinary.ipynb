```python
import pandas as pd
df = pd.read_csv('combined_filtered_genres.csv', index_col='name')
df.head()
```

```python
df['r-date'].dtype
```

```python
df['r-date'] = pd.to_datetime(df['r-date'], format='%B %d, %Y')
min_date = df['r-date'].min()
# print(min_date)
df['release_date_numeric'] = (df['r-date'] - min_date).dt.days

# Check the dataframe
print(df.head())
```

```python
df2  = df.copy()
df2['indie'] = df2['Combined_Genres'].str.contains('Indie', case=False).astype(int)
df2.head()
df.to_csv('all_numerical_only_indie.csv', index=True)
```

```python
df3  = df.copy()
df2['indie'] = df2['Combined_Genres'].str.contains('Indie', case=False).astype(int)
df2.head()
df.to_csv('all_numerical_only_indie.csv', index=True)
```

```python
df['genres'] = df['Combined_Genres'].str.split(',')

# Step 2: Get all unique genres
all_genres = set([genre for sublist in df['genres'] for genre in sublist])

# Step 3: Create binary columns for each genre
for genre in all_genres:
    df[genre] = df['genres'].apply(lambda x: 1 if genre in x else 0)

# Drop the 'combined_genres' and 'genres' columns if you no longer need them
df.drop(columns=['Combined_Genres', 'genres','r-date'], inplace=True)

# Print the resulting dataframe
print(df.head(1))
```
